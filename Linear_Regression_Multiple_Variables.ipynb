```python
import pandas as pd
import numpy  as np
# sklearn standfor Science Kit Learn
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/homepricesmulti.csv')
df
```

```python
#replace the NAN value with median
df.bedrooms.median()
```

```python
df.bedrooms.fillna(df.bedrooms.median(),inplace=True)
df
```

```python
x= df.drop('price' ,axis='columns')
x
```

```python
y=df.price
y
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

**y = mx1 + m2x2 + m3x3 + c **

```python
model.coef_
```

```python
model.intercept_
```

Find the price of home with 3000 sqr ft area ,3 bedrooms, 40 years old

```python
model.predict([[3000,3,40]])
```
