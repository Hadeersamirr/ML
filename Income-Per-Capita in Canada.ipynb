```python
import pandas as pd
import numpy  as np
# sklearn standfor Science Kit Learn
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
df
```

```python
df.head()
```

```python
df.rename(columns={'per capita income (US$)': 'income'}, inplace=True)
df
```

```python
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year , df.income ,color='red' , marker='*')
plt.show()
```

```python
year = df.drop('income',axis='columns')
print(type(year))
year
```

```python
income=df.drop('year' , axis ='columns')
print(type(income))
income
```

```python
reg= linear_model.LinearRegression()
reg.fit(year,income)
```

```python
 reg.predict([[2020]])
```

```python
reg.score(year,income)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year , df.income ,color='red' , marker='*')
plt.plot(df.year, reg.predict(df[['year']]),color='blue')
```
